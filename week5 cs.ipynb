```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as seabornInstance 
from sklearn.model_selection import train_test_split 
from sklearn import metrics
from  matplotlib import pyplot 
from  sklearn.linear_model import LogisticRegression
%matplotlib inline
```

```python
df=pd.read_csv(r'C:\Users\user\Desktop\Week 5\Python\heart.csv')
df
```

```python
df.info()
```

```python
pp=df.loc[:,['cp','ca','thal']]
pp
```

```python
ss = pp.astype('str')
```

```python
tt=pd.get_dummies(ss,drop_first=True)
tt
```

```python
df.drop(['cp','ca','thal'],axis=1,inplace = True)
df
```

```python
df = pd.concat([df,tt],axis=1)
```

```python
df
```

```python
X=df.drop('target',axis=1)
y=df['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=1)
```

```python
log_regression=LogisticRegression(max_iter=50000)
log_regression
```

```python
log_regression.fit(X_train,y_train)
```

```python
y_pred=log_regression.predict(X_test)
y_pred
```

```python
print(metrics.classification_report(y_test,y_pred))
```
